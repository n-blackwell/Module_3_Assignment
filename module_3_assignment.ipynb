```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
df = pd.read_csv('spotify.csv')
df.fillna(0, inplace=True)
df.set_index(pd.Series(df['song_id']), inplace=True, drop=True)
df.drop(['song_id'], axis=1, inplace=True)
df
```

```python
#6910 through around 6914 have null values. i switched them to 0
audio = df.drop(['uri', 'rank', 'artist_names', 'artists_num', 'artist_individual', 'artist_id', 'artist_genre', 'artist_img', 'collab', 'track_name', 'release_date', 'country', 'region', 'language', 'pivot', 'album_num_tracks', 'album_cover', 'source', 'peak_rank', 'previous_rank', 'weeks_on_chart', 'streams', 'week'], axis=1)
audio.fillna(0)
audio
```

```python
x = audio.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
audio = pd.DataFrame(x_scaled, columns= audio.columns, index=audio.index)
audio
```

```python
targets = [audio.loc[32904], audio.loc[29587], audio.loc[40669]]
distance_log = []
```

```python
import scipy.spatial.distance as sp
```

```python
x = 0
for t in targets:
    distance_log.append(1 - sp.cdist(audio, [t], metric="cosine")[:,0])

export1 = df
export1['Distances'] = distance_log[0]
export1.to_csv('lumberjack_with_similiarity.csv')

export2 = df
export2['Distances'] = distance_log[1]
export2.to_csv('tliktb_with_similiarity.csv')

export3 = df
export3['Distances'] = distance_log[2]
export3.to_csv('sotw_with_similiarity.csv')
```

```python
paired_distance_log = []
paired_distance_log.append(list(zip(df.index, distance_log[0])))
paired_distance_log.append(list(zip(df.index, distance_log[1])))
paired_distance_log.append(list(zip(df.index, distance_log[2])))
for paired_distance in paired_distance_log:
    print()
    for pair in sorted(paired_distance, key=lambda x: x[1], reverse=True)[:11]:
        song = df.loc[pair[0]]
        print(f'{song[9]} by {song[2]}: {pair[1]}')
```
